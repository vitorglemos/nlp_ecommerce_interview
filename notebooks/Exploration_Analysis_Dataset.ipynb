```python
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt

%config InlineBackend.figure_format = 'retina'

df = pd.read_csv("/content/gdrive/MyDrive/b2w/B2W-Reviews01.csv")
```

```python
df.head(5).T
```

```python
df.info()
```

```python
plt.xlabel("Count")
plt.ylabel("Rating")
plt.title("Balanced Rating Score by Users")
df.overall_rating.value_counts().plot(kind='barh')
```

```python
map_sentiment = {1: "negative",
                 2: "negative",
                 3: "neutral",
                 4: "positive",
                 5: "positive"}
df["overall_rating"] = df["overall_rating"].map(map_sentiment)
plt.xlabel("Count")
plt.ylabel("Sentiment")
plt.title("Sentiment target in dataset")
df.overall_rating.value_counts().plot(kind='barh')
```

```python
sns.countplot(x="reviewer_gender", hue="overall_rating", data=df)
plt.title("Count Reviewer by Gender")
```

```python
sns.countplot(x="recommend_to_a_friend", hue="overall_rating", data=df)
plt.title("Recomend to friend based in rating")
```
